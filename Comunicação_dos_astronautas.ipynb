```julia
# -*- coding: utf-8 -*-
# Simulao simplificada da comunicao Apollo 11
# Autor: Gabriel Xavier Oscar

import networkx as nx
import matplotlib.pyplot as plt

# Criando a rede de comunicao
G = nx.DiGraph()

# Ns principais
nodes = [
    "Astronautas (LM e CSM)",
    "S-Band Transmission",
    "Deep Space Network",
    "Houston - CAPCOM",
    "Engenheiros / Computadores NASA"
]
G.add_nodes_from(nodes)

# Ligaes
edges = [
    ("Astronautas (LM e CSM)", "S-Band Transmission"),
    ("S-Band Transmission", "Deep Space Network"),
    ("Deep Space Network", "Houston - CAPCOM"),
    ("Houston - CAPCOM", "Engenheiros / Computadores NASA")
]
G.add_edges_from(edges)

# Visualizao
plt.figure(figsize=(10,6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=2500, font_size=10, arrows=True)
plt.title("Fluxo de Comunicao  Apollo 11", fontsize=14)
plt.show()
```

```julia
import networkx as nx
import matplotlib.pyplot as plt

# Criando o grafo dirigido
G = nx.DiGraph()

# Ns principais
nodes = [
    "Astronautas (LM e CSM)",
    "S-Band Transmission",
    "Deep Space Network",
    "Houston  CAPCOM",
    "Engenheiros / Computadores"
]
G.add_nodes_from(nodes)

# Ligaes (fluxo da comunicao)
edges = [
    ("Astronautas (LM e CSM)", "S-Band Transmission"),
    ("S-Band Transmission", "Deep Space Network"),
    ("Deep Space Network", "Houston  CAPCOM"),
    ("Houston  CAPCOM", "Engenheiros / Computadores")
]
G.add_edges_from(edges)

# Visualizao
plt.figure(figsize=(10,6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color="skyblue", node_size=2500,
        arrowsize=20, font_size=10, font_weight="bold", edge_color="gray")
plt.title("Fluxo de Comunicao  Apollo 11", fontsize=14, fontweight="bold")
plt.show()
```
